# Funções

```python
def Sexo_(sex):
    switcher = {
        1: "Homem",
        2: "Mulher",
    }
    return switcher.get(sex, "SemResp")


def Idade_(ida):
    switcher = {
    1:"Menos de 13 anos",
    2:"13 a 15 anos",
    3:"16 ou 17 anos",
    4:"18 anos ou mais"   
    }
    return switcher.get(ida, "SemResp")


def Etnia_(etnia):
    switcher = {
        1: "Branca",
        2: "Preta",
        3: "Amarela",
        4: "Parda",
        5: "Indígena",
        -2:"Abandono"
    }
    return switcher.get(etnia, "SemResp")


def AnoEscolar_(ano):
    switcher = {
        1: "Ano 6 Fund",
        2: "Ano 7 Fund",
        3: "Ano 8 Fund",
        4: "Ano 9 Fund",
        5: "Ano 1 Med",
        6: "Ano 2 Med",
        7: "Ano 3 Med",
        -2:"Abandono"
    }
    return switcher.get(ano, "SemResp")

def QuandoTerminarFundMed(quero):
    switcher = {
        
        -1:"Pulo do questionário",
        1:"Somente continuar estudando",
        2:"Somente trabalhar",
        3:"Continuar estudando e trabalhar",
        4:"Seguir outro plano",
        5:"Não sei"
    }
    return switcher.get(quero, "SemResp")


def moraPaiMae(resp):
    switcher={
        -2: "Abandono",
        1:"Sim",
        2:"Não"
    }
    return switcher.get(resp, "SemResp")


def PossuiCel_(resp):
    switcher={
        -2: "Abandono",
        1:"Sim",
        2:"Não"
    }
    return switcher.get(resp, "SemResp")


def PossuiComp_(resp):
    switcher={
        -2: "Abandono",
        1:"Sim",
        2:"Não"
    }
    return switcher.get(resp, "SemResp")


def InternetCasa_(resp):
    switcher={
        -2: "Abandono",
        1:"Sim",
        2:"Não"
    }
    return switcher.get(resp, "SemResp")

def GrauEsinoMae_(resp):
    switcher={
        -2:"Abandono de questionário",
        1:"Minha mãe não estudou",
        2:"Não terminou o Ensino Fundamental (1º grau)",
        3:"Terminou o Ensino Fundamental (1º grau)",
        4:"Não terminou o Ensino Médio (2º grau)",
        5:"Terminou o Ensino Médio (2º grau)",
        6:"Não terminou o Ensino Superior (faculdade)",
        7:"Terminou o Ensino Superior (faculdade)",
        8:"Não sei"
        
    }
    return switcher.get(resp, "SemResp")


def REGIAO_(resp):
    switcher={
    1:"Norte",
    2:"Nordeste",
    3:"Sudeste",
    4:"Sul",
    5:"Centro-Oeste"

    }
    return switcher.get(resp, "SemResp")


def SITUACAO_(resp):
    switcher={
        1:"Urbana",
        2:"Rural"
    }
    return switcher.get(resp, "SemResp")


def DEP_ADMIN_(resp):
    switcher={
        1:"Pública",
        2:"Privada"
    }
    return switcher.get(resp, "SemResp")



def EnsinoDistancia_(PossuiComp,InternetCasa ):
    if PossuiComp=="Sim" and InternetCasa=="Sim":
        retorno="PossuiCompComInternet"
    elif PossuiComp=="Sim" and InternetCasa=="Não":
        retorno="PossuiCompSemInternet"
    elif PossuiComp=="Não" and InternetCasa=="Sim":
        retorno="PossuiInternetSemComp"
    elif PossuiComp=="Não" and InternetCasa=="Não":
        retorno="NemCompNemInternet"   
    else:
        retorno="NãoInformado"
    
    return retorno
```
